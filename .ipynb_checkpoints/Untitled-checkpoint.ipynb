```python
import cv2
import mediapipe as mp
import numpy as np
mp_drawing = mp.solutions.drawing_utils
mp_pose = mp.solutions.pose
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    cv2.imshow('Mediapipe Feed', frame)
    
    #if para salir del ciclo de la camara, caso de que se presione la letra q o que se presione la equis de salir
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

# Make detections

```python
cap = cv2.VideoCapture(0)
##instancia de mediapipe
with mp_pose.Pose(min_detection_confidence = 0.5, min_tracking_confidence = 0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()

        #recoloreando la imagen
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        #procesamos las imagenes  y detectamos
        results = pose.process(image)

        #recoloreamos nuevamente a BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        #renderizamos las detecciones
        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)
        

        #creamos la camara
        cv2.imshow('Mediapipe Feed', image)

        #if para salir del ciclo de la camara, caso de que se presione la letra q o que se presione la equis de salir
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
        
cap.release()
cv2.destroyAllWindows()
```
